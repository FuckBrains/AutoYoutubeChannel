```python
#get SubReddit, submission & comments
subreddit=reddit.subreddit('AskReddit')
hot_posts=subreddit.hot(limit=1)
commenttext=''
replies=[]
title=''
print(subreddit.id)
for submission in hot_posts:
    print(100*'*')
    submission=reddit.submission(id='hpp1w8')
    print(submission.title)
    titleTranslated=translate_clean.translateToSpanish(submission.title)
    tts=gTTS(titleTranslated,lang='es')
    tts.save('videoEditor/title.mp3')
    title=titleTranslated
    submission.comments_sort='best'
    comments = submission.comments.list()
    index=0
    for comment in comments:
        if isinstance(comment,MoreComments):
            break
        if index==25:
            break
        print(10*'_')
        commenttext=comment
        print(comment.body)#filter by len, so bigger comments are extracted no comments with 2 o r 3 words
        try:
            replyTranslated=translate_clean.translateToSpanish(commenttext.body)
            tts=gTTS(replyTranslated,lang='es')
            tts.save('videoEditor/comment'+str(index)+'.mp3')
        except:
            continue
        linesOfReply=translate_clean.separateByLines(text=replyTranslated,charsInLine=145)
        reply={'author':comment.author.name,'replyLines':linesOfReply}
        replies.append(reply)
        index+=1
        
replyTranslated=translate_clean.translateToSpanish(commenttext.body)
print(replyTranslated)
listLines=translate_clean.separateByLines(text=replyTranslated,charsInLine=145)
print(listLines)
```
